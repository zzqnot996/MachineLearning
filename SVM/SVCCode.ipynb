````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, svm

iris = datasets.load_iris()

x = iris.data[:, :2]
y = iris.target
h = .05
#-----------------------------------------------------------------------------------------------------------
# 用线性内核
# 用非线性内核，比如多项式内核，生成非线性决策边界
# 使用rbf径向基内核生成的径向决策区域
#-----------------------------------------------------------------------------------------------------------
# svc = svm.SVC(kernel='linear',C=1.0,degree=3).fit(x,y)
# svc = svm.SVC(kernel='poly',C=1.0,degree=3).fit(x,y)
svc = svm.SVC(kernel='rbf',C=1.0,degree=3).fit(x,y)

# #-----------------------------------------------------------------------------------------------------------
# 生成网格limit
#-----------------------------------------------------------------------------------------------------------
x_min, x_max = x[:, 0].min() - .5, x[:, ].max() + .5
y_min, y_max = x[:, 1].min() - .5, x[:, 1].max() + .5
h =.02

X,Y = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min,y_max, h))
Z = svc.predict(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)
#-----------------------------------------------------------------------------------------------------------
# https://blog.csdn.net/lens___/article/details/83960810
#-----------------------------------------------------------------------------------------------------------
plt.contourf(X,Y,Z,alpha=0.4)
plt.contour(X,Y,Z,colors='k')
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()
````

SVC算法对更复杂的数据集的分类情况

扩展到三个类别，因为Iris数据集包含三个类别的鸢尾花

对于这个数据集，决策边界相互交叉，把决策空间（2D）和决策体（3D）分成两个部分

两个线性模型均有线性决策边界（相交的超平面），而使用非线性内核的模型（多项式或高斯RBF）有非线性决策边界，后者在处理依赖于内核和参数的数据时更加灵活

`svm.SVC`是`sklearn`库中的一个函数，用于实现支持向量机（Support Vector Machines，SVM）。以下是`svm.SVC`的一些主要参数的详解：

```python
svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, decision_function_shape=None,random_state=None)
```

- `C`：C-SVC的惩罚参数C，默认值是1.0。C越大，相当于惩罚松弛变量，希望松弛变量接近0，即对误分类的惩罚增大，趋向于对训练集全分对的情况，这样对训练集测试时准确率很高，但泛化能力弱。C值小，对误分类的惩罚减小，允许容错，将他们当成噪声点，泛化能力较强¹。
- `kernel`：核函数，默认是rbf，可以是‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’¹。
- `degree`：多项式poly函数的维度，默认是3，选择其他核函数时会被忽略¹。
- `gamma`： ‘rbf’,‘poly’ 和‘sigmoid’的核函数参数。 默认是’auto’，则会选择1/n_features¹。
- `coef0`：核函数的常数项。 对于‘poly’和 ‘sigmoid’有用¹。
- `probability`：是否采用概率估计，默认为False¹。
- `shrinking`：是否采用shrinking heuristic方法，默认为true¹。
- `tol`：停止训练的误差值大小，默认为1e-3¹。
- `cache_size`：核函数cache缓存大小，默认为200¹。
- `class_weight`：类别的权重，字典形式传递¹。
- `verbose`：允许冗余输出¹。
- `max_iter`：最大迭代次数。 -1为无限制¹。
- `decision_function_shape`：‘ovo’, ‘ovr’ or None, default=None¹。
- `random_state`：数据洗牌时的种子值，int值¹。
